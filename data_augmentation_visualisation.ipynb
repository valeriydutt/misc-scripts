```python
import numpy as np
from skimage.io import imread, imsave
import matplotlib.pyplot as plt
from skimage import transform
from skimage.transform import rotate, AffineTransform
from skimage.util import random_noise
from skimage.filters import gaussian
from scipy import ndimage
```

```python

def plot_side_by_side(img1, img2, title1, title2, cmap = None):
    fig = plt.figure(tight_layout='auto', figsize=(10, 7))
    fig.add_subplot(221)
    plt.title(title1)
    plt.imshow(img)

    fig.add_subplot(222)
    plt.title(title2)
    plt.imshow(img2, cmap = None)
    return fig

```

```python
# load Image
# dividing by 255 to scale intensity values between 0,1
img = imread('C:/Users/Owner/Desktop/dissertation/data/train/13_left.jpeg') / 255
```

```python
# plot original Image
plt.imshow(img)
plt.show()
```

```python
# image rotation using skimage.transformation.rotate
rotate30 = rotate(img, angle=30)
rotate45 = rotate(img, angle=45)
rotate60 = rotate(img, angle=60)
rotate90 = rotate(img, angle=90)

fig = plt.figure(tight_layout='auto', figsize=(10, 7))

fig.add_subplot(221)
plt.title('30° Rotation')
plt.imshow(rotate30)

fig.add_subplot(222)
plt.title('45° Rotation')
plt.imshow(rotate45)

fig.add_subplot(223)
plt.title('60° Rotation')
plt.imshow(rotate60)

fig.add_subplot(224)
plt.title('90° Rotation')
plt.imshow(rotate90)

plt.show()
```

```python
# image shearing using sklearn.transform.AffineTransform
# try out with differnt values of shear 
tf = AffineTransform(shear=0.5)
sheared = transform.warp(img, tf, order=1, preserve_range=True, mode='wrap')

sheared_fig = plot_side_by_side(img, sheared, 'Original', 'Sheared')
```

```python

# Image rescaling with sklearn.transform.rescale
rescaled = transform.rescale(img, 1.05)

rescaled_fig = plot_side_by_side(img, rescaled, 'Original', 'Rescaled')
plt.show()

print('Original Shape: ',img.shape)
print('Rescaled Shape: ',rescaled.shape)
```

```python
# flip up-down using np.flipud
up_down = np.flipud(img)

fig_updown = plot_side_by_side(img, up_down, 'Original', 'Vertical Flip')
plt.show()
```

```python
# flip up-down using np.flipud
left_right = np.fliplr(img)

fig_lr = plot_side_by_side(img, left_right, 'Original', 'Horizontal Flip')
plt.show()
```

```python
# Apply Random Noise to image using skimage.utils.random_noise
noised = random_noise(img, mode='gaussian', var=0.4**2, clip=True)

fig_noised = plot_side_by_side(img, noised, 'Original', 'Gaussian Noise')
plt.show()
```

```python
# Increasing the brighness of the Image
# Note: Here we add 100/255 since we scaled Intensity values of Image when loading (by dividing it 255)
highB = img + (100/255)

fig_highB = plot_side_by_side(img, highB, 'Original', 'High Brightness')
plt.show()
```

```python
# Increasing the contrast of the Image
# Note: Here we add 100/255 since we scaled Intensity values of Image when loading (by dividing it 255)
highC = img * 1.5

fig_highB = plot_side_by_side(img, highC, 'Original', 'High Contrast')
plt.show()
```
